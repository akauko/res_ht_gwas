# Run lift over pipeline

This script runs liftover_plink_beds pipeline: https://github.com/dnanexus-rnd/liftover_plink_beds
We try now run bash from ttyd app at UKB RAP.

Requirements:

* dx-toolkit
* dxCompiler
* Java
* Chain file
* Reference genome


```bash
apt install default-jre
wget https://github.com/dnanexus/dxCompiler/releases/download/2.11.4/dxCompiler-2.11.4.jar
git clone https://github.com/dnanexus-rnd/liftover_plink_beds.git
```

**Compile wdl at bash app "ttyd"**

Note that if compiler is run at the local computer, project ID must be defined explicitely. I could use environmental variable, because I usedapp ttyd in the ukb RAP environment.

```bash
wdl="liftover_plink_beds/liftover_plink_beds.wdl"
folder="/liftover_plink_beds/workflow/"
java -jar dxCompiler-2.11.4.jar compile $wdl -project $DX_PROJECT_CONTEXT_ID -folder $folder
```

**Excecute workflow at RAP**

```bash
json="liftover_input.json"
dx download /liftover_plink_beds/$json
dx run workflow-Gb5g098JFFQb3Qx26jk825Pq -f $json --brief -y --priority normal

#analysis-Gb612ZQJFFQzZXFPXx068FFJ
```
